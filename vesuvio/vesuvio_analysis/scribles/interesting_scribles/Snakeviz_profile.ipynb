```python
import numpy as np

def load_ip_file(spectrum):
    """Instrument parameters of VESUVIO"""  
    
    #print "Loading parameters from file: ", namedtuple
    ipfile = r'C:\Users\guijo\Desktop\Work\ip2018.par'
    f = open(ipfile, 'r')
    data = f.read()
    lines = data.split('\n')
    for line in lines:       
        col = line.split('\t')
        if col[0].isdigit() and int(col[0]) == spectrum:
            angle = float(col[2])
            T0 = float(col[3])
            L0 = float(col[4])
            L1 = float(col[5])
    f.close()
    return angle, T0, L0, L1    
    
 
def load_ip_file_improved(spectrum):
    
    ipfile = r'C:\Users\guijo\Desktop\Work\ip2018.par'
    f = open(ipfile, 'r')
    data = f.read()
    lines = data.split("\n")[1:-1]  #take out the first line of non numbers and the last empty line
    data = list(map(lambda line: list(map(float, line.split("\t"))), lines))   #map is faster than for loops, list comprehension faster than map
    data = np.array(data)
    row = data[data[:, 0]==spectrum]
    angle, T0, L0, L1 = row[0, 2:]   #0 because its an array inside of an array
    return angle, T0, L0, L1

def load_ip_file_improved_2(spectrum):  
    ipfile = r'C:\Users\guijo\Desktop\Work\ip2018.par'
    f = open(ipfile, 'r')
    data = f.read()
    f.close()
    lines = data.split("\n")[1:-1]  #take out the first line of non numbers and the last empty line
    data_str = [line.split("\t") for line in lines]
    data = [[float(i) for i in line] for line in data_str]
    data = np.array(data)
    row = data[data[:, 0]==spectrum]
    angle, T0, L0, L1 = row[0, 2:]   #0 because its an array inside of an array
    return angle, T0, L0, L1
    
```

```python
import cProfile, pstats
profiler = cProfile.Profile()
profiler.enable()
load_ip_file(150)
profiler.disable()
stats = pstats.Stats(profiler).sort_stats('cumtime')
stats.print_stats()
```

```python
#Example of a function that is now optimized, runs about 4 times faster

from mantid.simpleapi import *
import numpy as np

def normalise_workspace(ws_name):
    tmp_norm = Integration(ws_name)
    Divide(LHSWorkspace=ws_name,RHSWorkspace="tmp_norm",OutputWorkspace=ws_name)
    DeleteWorkspace("tmp_norm")
    
def convert_to_y_space_and_symmetrise_OPTIMIZED(ws_name,mass):  #Need to go over it
    """input: TOF workspace
       output: workspace in y-space for given mass with Ydata symetrised"""
          
    ws_y, ws_q = ConvertToYSpace(InputWorkspace=ws_name,Mass=mass,OutputWorkspace=ws_name+"_JoY",QWorkspace=ws_name+"_Q")
    max_Y = np.ceil(2.5*mass+27)    #where from
    rebin_parameters = str(-max_Y)+","+str(2.*max_Y/120)+","+str(max_Y)   #first bin boundary, width, last bin boundary, so 120 bins over range
    ws_y = Rebin(InputWorkspace=ws_y, Params = rebin_parameters, FullBinsOnly=True, OutputWorkspace=ws_name+"_JoY")
   
    matrix_Y = np.zeros((ws_y.getNumberHistograms(), ws_y.blocksize()))
    for spec_idx in range(len(matrix_Y)):                 #pass the y-data onto an array to easily manipulate
        matrix_Y[spec_idx, :] = ws_y.readY(spec_idx)        
    matrix_Y[matrix_Y != 0] = 1
    sum_Y = np.nansum(matrix_Y, axis=0)   
    ws_y = SumSpectra(InputWorkspace=ws_y, OutputWorkspace=ws_name+"_JoY")
    tmp=CloneWorkspace(InputWorkspace=ws_y)
    tmp.dataY(0)[:] = sum_Y
    tmp.dataE(0)[:] = np.zeros(tmp.blocksize())
    
    ws = Divide(LHSWorkspace=ws_y, RHSWorkspace=tmp, OutputWorkspace =ws_name+"_JoY")
    ws.dataY(0)[:] = (ws.readY(0)[:] + np.flip(ws.readY(0)[:])) / 2           #symetrise dataY
    ws.dataE(0)[:] = (ws.readE(0)[:] + np.flip(ws.readE(0)[:])) / 2           #sumetrise dataE
    normalise_workspace(ws)
    return max_Y 

def convert_to_y_space_and_symmetrise_ORIGINAL(ws_name,mass):  #Need to go over it       
    ws_y, ws_q = ConvertToYSpace(InputWorkspace=ws_name,Mass=mass,OutputWorkspace=ws_name+"_JoY",QWorkspace=ws_name+"_Q")    
    max_Y = np.ceil(2.5*mass+27)    #where from
    rebin_parameters = str(-max_Y)+","+str(2.*max_Y/120)+","+str(max_Y)   #first bin boundary, width, last bin boundary, so 120 bins over range
    ws = Rebin(InputWorkspace=ws_name+"_JoY", Params = rebin_parameters,FullBinsOnly=True, OutputWorkspace= ws_name+"_JoY")
    tmp=CloneWorkspace(InputWorkspace=ws_name+"_JoY")
    for j in range(tmp.getNumberHistograms()):
        for k in range(tmp.blocksize()):
            tmp.dataE(j)[k] =0.
            if (tmp.dataY(j)[k]!=0):
                tmp.dataY(j)[k] =1.
    tmp=SumSpectra('tmp')
    ws = SumSpectra(InputWorkspace=ws_name+"_JoY",OutputWorkspace=ws_name+"_JoY")
    ws = Divide(LHSWorkspace=ws_name+"_JoY", RHSWorkspace="tmp", OutputWorkspace =ws_name+"_JoY")
    ws=mtd[ws_name+"_JoY"]
    tmp=CloneWorkspace(InputWorkspace=ws_name+"_JoY")
    for k in range(tmp.blocksize()):
        tmp.dataE(0)[k] =(ws.dataE(0)[k]+ws.dataE(0)[ws.blocksize()-1-k])/2.
        tmp.dataY(0)[k] =(ws.dataY(0)[k]+ws.dataY(0)[ws.blocksize()-1-k])/2
    RenameWorkspace(InputWorkspace="tmp",OutputWorkspace=ws_name+"_JoY")
    normalise_workspace(ws_name+"_JoY")
    return max_Y

ws = Load("C:/Users/guijo/Desktop/Work/CePtGe12_backward_100K_scipy/CePtGe12_100K_DD_.nxs", OutputWorkspace="CePtGe12_100K_DD_.nxs")
```
